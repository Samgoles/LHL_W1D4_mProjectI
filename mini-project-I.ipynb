## Task 2
What are the different modes of transport which are operated by Transfer for London? How many of modes do they have?

Print the list with different modes of transport, plus their count. Example output:
```
[bus, cable-car,.....]
Number of different modes of transport is: xyz
```

We need to search the documentation for correct request.

````python
url2='https://api.tfl.gov.uk/StopPoint/Meta/Modes'
res2=re.get(url2+url_append)
if res2.status_code !=200:
    print('no valid request')
else:
    a2=res2.json()
    
a2
````

````python
modename=[]
modenumber=[]
for i in a2:
    if i['modeName'] not in modename:
        #print(i['modeName'])
        modename.append(i['modeName'])
        modenumber.append(1)
    else:
        modenumber[modename.index(i['modeName'])]+=1
    #   print(1)

print('the mode names are :\n'+str(modename)+'\n'+'the total modes number is  '+str(len(modename)))

    
````

## Task 3
How many BikePoints in London are operated by Transfor for London? How many docks are in **all** BikePoints? There is the information for empty and full docks for each BikePoint.

````python
url3='https://api.tfl.gov.uk/BikePoint'
res3=re.get(url3+url_append)
if res3.status_code!=200:
    print('not valid response')
else:
    a3=res3.json()
    a3
````

````python
NbBikes=0
NbEmptyDocks=0
NbDocks=0
for a in a3:
    NbBikes+=int(a['additionalProperties'][6]['value'])
    NbEmptyDocks+=int(a['additionalProperties'][7]['value'])
    NbDocks+=int(a['additionalProperties'][8]['value'])
````

````python
print(f'number of Bikepoints in londen is \n {len(a3)}')
print('Total number of docks is {} \n Total number of Bikes is {} \n Total number of Empty docks is {} \n Total number of broken Docks is {}\n which is obtained by Total number of docks -(Total number of Bikes+Total number of empty docks) '.format(NbDocks,NbBikes,NbEmptyDocks,NbDocks-(NbBikes+NbEmptyDocks)))
````
